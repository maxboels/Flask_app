```python
import jgraph

import json

with open('edges_convertcsv.json') as f:
  edges = json.load(f)

with open('nodes_convertcsv.json') as f:
  nodes = json.load(f)

# remove column.
for node in range(len(nodes)):
    nodes[node].pop('FIELD4', None)

dict_nodes = {}

for i in range(len(nodes)):

    new_key = str(nodes[i]['node_id'])

    nodes[i].pop('node_id', None)
    nodes[i].pop('node_color', None)
    nodes[i].pop('node_label', None)
     
    dict_nodes[new_key] = nodes[i]

new_edges = []

for i in range(len(edges)):
    temp = edges[i]
    temp.pop('weights', None)
    
    temp['source_id'] = str(temp['source_id'])
    
    temp['target_id'] = str(temp['target_id']) 
    
    new_edges.append(temp)
    

graph = {'nodes': dict_nodes, 'edges': new_edges}


print(graph)
# print(dict_nodes.keys())
# print(dict_nodes[3].values())

jgraph.draw(graph)
```
